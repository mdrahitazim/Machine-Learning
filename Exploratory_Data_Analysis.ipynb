# Plots

```python
import matplotlib.pyplot as plt
```

```python

temporal_trend = df.groupby('Year')['Estimated prevalence of TB (all forms) per 100 000 population'].sum()

# Plot the trend
plt.figure(figsize=(10, 6))
temporal_trend.plot()
plt.title('TB prevalence per year')
plt.xlabel('Year')
plt.ylabel('TB prevalence')
plt.grid(True)
plt.show()
```

```python
temporal_trend = df.groupby('Year')['Estimated total population number'].sum()

# Plot the trend
plt.figure(figsize=(10, 6))
temporal_trend.plot()
plt.title('Total Population per year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
temporal_trend = df.groupby('Year')['Estimated prevalence of TB (all forms)'].sum()

# Plot the trend
plt.figure(figsize=(10, 6))
temporal_trend.plot()
plt.title('Total Population per year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```
